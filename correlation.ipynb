```python
# renaming the date column : will help to merge datasets later
fire = fire.rename(columns = {'acq_date':'date'})
```

```python
# filtering out the data for latitude within 28-31 and longitude within 73-78
filter = fire['longitude'] > 73
fire = fire[filter]
filter = fire['longitude'] < 78
fire = fire[filter]
filter = fire['latitude'] > 28
fire = fire[filter]
filter = fire['latitude'] < 31 
fire = fire[filter]

filter = pm2p5['longitude'] > 73
pm2p5 = pm2p5[filter]
filter = pm2p5['longitude'] < 78
pm2p5 = pm2p5[filter]
filter = pm2p5['latitude'] > 28
pm2p5 = pm2p5[filter]
filter = pm2p5['latitude'] < 31 
pm2p5 = pm2p5[filter]

filter = bcaod550['longitude'] > 73
bcaod550 = bcaod550[filter]
filter = bcaod550['longitude'] < 78
bcaod550 = bcaod550[filter]
filter = bcaod550['latitude'] > 28
bcaod550 = bcaod550[filter]
filter = bcaod550['latitude'] < 31
bcaod550 = bcaod550[filter]

filter = pm10['longitude'] > 73
pm10 = pm10[filter]
filter = pm10['longitude'] < 78
pm10 = pm10[filter]
filter = pm10['latitude'] > 28
pm10 = pm10[filter]
filter = pm10['latitude'] < 31
pm10 = pm10[filter]

filter = tcco['longitude'] > 73
tcco = tcco[filter]
filter = tcco['longitude'] < 78
tcco = tcco[filter]
filter = tcco['latitude'] > 28
tcco = tcco[filter]
filter = tcco['latitude'] < 31
tcco = tcco[filter]

filter = tcso2['longitude'] > 73
tcso2 = tcso2[filter]
filter = tcso2['longitude'] < 78
tcso2 = tcso2[filter]
filter = tcso2['latitude'] > 28
tcso2 = tcso2[filter]
filter = tcso2['latitude'] < 31
tcso2 = tcso2[filter]
```

```python
# extracting date from datetime and creating another date column with it
pm2p5['time'] = pd.to_datetime(pm2p5.time)
pm2p5['date'] = pm2p5['time'].dt.date

bcaod550['time'] = pd.to_datetime(bcaod550.time)
bcaod550['date'] = bcaod550['time'].dt.date

pm10['time'] = pd.to_datetime(pm10.time)
pm10['date'] = pm10['time'].dt.date

tcco['time'] = pd.to_datetime(tcco.time)
tcco['date'] = tcco['time'].dt.date

tcso2['time'] = pd.to_datetime(tcso2.time)
tcso2['date'] = tcso2['time'].dt.date
```

```python
# selecting the variables from multiple dataframes that we need for the correlation matrix
m1 = fire[['date','frp','brightness', 'bright_t31']]
m2 = pm2p5[['date','pm2p5']]
m3 = bcaod550[['date','bcaod550']]
m4 = pm10[['date','pm10']]
m5 = tcco[['date','tcco']]
m6 = tcso2[['date','tcso2']]
```

```python
# grouping the data on per day basis using mean
m1.date = pd.to_datetime(m1.date)
m1 = m1.groupby(pd.Grouper(key='date', freq='1D')).mean()

m2.date = pd.to_datetime(m2.date)
m2 = m2.groupby(pd.Grouper(key='date', freq='1D')).mean()

m3.date = pd.to_datetime(m3.date)
m3 = m3.groupby(pd.Grouper(key='date', freq='1D')).mean()

m4.date = pd.to_datetime(m4.date)
m4 = m4.groupby(pd.Grouper(key='date', freq='1D')).mean()

m5.date = pd.to_datetime(m5.date)
m5 = m5.groupby(pd.Grouper(key='date', freq='1D')).mean()

m6.date = pd.to_datetime(m6.date)
m6 = m6.groupby(pd.Grouper(key='date', freq='1D')).mean()
```

```python
# merging various dataframes into one using the date variable
df = pd.merge(m1, m2, on='date')
df = pd.merge(df, m3, on='date')
df = pd.merge(df, m4, on='date')
df = pd.merge(df, m5, on='date')
df = pd.merge(df, m6, on='date')
```
